# FORECASTING GDP

We want to understand how much of an impact GDP has on the stock market and if we are able to make stock predictions based on the trend of the GDP. To do this, we need to forecast what the GDP will be for the upcoming quarters and understand the trend of the market. We will using data from the FRED website that has FRED-MD (monthly data) and FRED-QD (quarterly data) datasets. Below is the url for the data:

https://research.stlouisfed.org/econ/mccracken/fred-databases/

This website contains the databases with macroeconomic data. We will be using the the FRED-MD dataset (this dataset contains the indicators used to forecast GDP) and only the real GDP from the FRED-QD dataset.

We are going to use two methods forecast GDP. First, we are going to use a Dynamic Factor Model that allows us to use a small number of unobserved "features" to explain the GDP. A large benefit of using this model is that we can forcast GDP, which is observed quarterly, with indicators that are observed monthly. Second, we are going to run a time series analysis on the dataset. We are going to use multiple machine learning models and determine which model predicted GDP the best.

Once we are able to get an accurate forcast of the GDP for the upcoming quarters, we will be able to make stock predictions and invest accordingly.

## Transform Data (Dynamic Factor Model)

Now that we have collected and clean all of our data, we must create a copy of our FRED-MD dataset with the transformed data. We have to transform the FRED-MD dataset because we need the features to be stationary. When forecasting the future, our Dynamic Factor Model assumes that each point is independent of one another. The best way to model this is to have our features be stationary.

We transformed the dataset according to the appendix provided from the FRED website.

```python
def transform(indicator, transforms):
    transformation = transforms[indicator.name]
    
    # 1 -> No Transformation
    if transformation == 1:
        pass
    # 2 -> Difference
    elif transformation == 2:
        indicator = indicator.diff()
    # 3 -> Difference Squared
    elif transformation == 3:
        indicator = indicator.diff().diff()
    # 4 -> Log
    elif transformation == 4:
        indicator = np.log(indicator)
    # 5 -> Log Difference
    elif transformation == 5:
        indicator = np.log(indicator).diff()
    # 6 -> Log Difference Squared
    elif transformation == 6:
        indicator = np.log(indicator).diff().diff()
    # 7 -> Exact Difference
    elif transformation == 7:
        indicator = ((indicator / indicator.shift(1)) - 1)
    
    return indicator
```
